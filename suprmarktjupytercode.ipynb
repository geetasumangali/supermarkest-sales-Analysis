```python
plt.figure(figsize=(20,8))
sb.lineplot(x='Date',y='Total',data=dt,color='green',errorbar=('ci',False))
plt.xticks(rotation='vertical')
plt.title('Plot - Date v/s Number of Sales',size=20)
plt.xlabel('Date',size=15)
plt.ylabel('SALES',size=12)
plt.show()
```

```python
plt.figure(figsize=(20,8))
sb.lineplot(x='Date',y='Total',data=dt,color='g',errorbar=('ci',True),hue='Branch',style='Branch')
plt.xticks(rotation='vertical')
plt.title('BRANCH-WISE: DATE v/s NUMBER OF SALES',size=20)
plt.xlabel('Date',size=15)
plt.ylabel('SALES',size=12)
plt.show()
```

```python
sb.countplot(x='Product line',data=dt)
plt.title('Products Sold(Mean) - Category Wise',size=15)
plt.xlabel('Category',size=15)
plt.ylabel('Count',size=12)
plt.xticks(rotation='vertical')
plt.show()
```

```python
sb.countplot(x='Payment',data=dt)
plt.title('Mean Purchase by Different Payment Modes',size=15)
plt.xlabel('Payment Mode',size=15)
plt.ylabel('Number of Purchases',size=12)
plt.xticks(rotation='vertical')
plt.show()
```

```python
sb.barplot(x='Branch',y="Total",data=dt,hue='Product line',errorbar=('ci',False))
plt.title('Branch Wise-Average Sale')
plt.grid()
plt.show()
```

```python
sales=dt.groupby('Date')['Total'].sum().reset_index()
sales
```

```python
plt.figure(figsize=(20,5))
sb.lineplot(x='Date',y='Total',data=sales,color='g')
plt.title('Sales Trend')
plt.ylabel('Sales($)')
plt.xticks(rotation='vertical')
plt.show()
```

```python
rev=dt.groupby('Branch').sum(numeric_only=True).reset_index()
rev
```

```python
sb.barplot(x='Branch',y='Total',data=rev)
plt.ylabel("Revenue in $")
plt.title('Total Revene')
plt.show()
```

```python
sb.barplot(x='Branch',y='gross income',data=rev,color='b')
plt.ylabel("Income in $")
plt.title('NET PROFIT')
plt.show()
```

```python
dt['hour']=dt['Time'].apply(lambda x: x.split(':')[0])
hour=dt.groupby('hour')['Total'].sum().reset_index().sort_values(by='hour')
plt.figure(figsize=(10,5))
sb.lineplot(x=hour['hour'],y=hour['Total'])
plt.title('Plot Showing Sales from 10am - 8pm')
plt.xlabel('Time(24Hr-Format)')
plt.ylabel('Sales($)')
plt.show()
```

```python
pie=dt.groupby('Product line')['Total'].sum().reset_index()
slices=list(pie['Total'])
name=list(pie['Product line'])
print(slices, name)
```
